# Convert Confusion Matrix to JSON

Previously, in the reports, I stored the confusion matrix in raw format into the report.

In later iterations, I stored them in json format so they fit inside the csv file better

In this notebook, I will convert old reports and restore them in json format

smote_time_min - this should be changed to sampling_time_min

```python
import pandas as pd
import json
import logging
import ast

logging.basicConfig(level=logging.INFO)

log = logging.getLogger(__name__)
```

```python
REPORT_DIR = "../../reports"
REPORT_FILE = f'{REPORT_DIR}/2019-11-01-run_classifiers-50k-report.csv'
report_init = pd.read_csv(REPORT_FILE)
```

```python
import re

def convert_string_to_array(cm_raw: str):
    if isinstance(cm_raw, str) and len(cm_raw) > 0:
        cm_raw = cm_raw.replace("\n", "")
        cm_raw = re.sub(r"\]\ +\[", "], [", cm_raw)
        cm_raw = re.sub(r'(\d+)\s+(\d+)', r'\1, \2', cm_raw)
        # have to do this twice as the first time it will not work
        cm_raw = re.sub(r'(\d+)\s+(\d+)', r'\1, \2', cm_raw)
        log.debug(cm_raw)
        cm_array = ast.literal_eval(cm_raw)
        return cm_array
    else:
        return None


def convert_to_json(cm_raw):
    cm_array = convert_string_to_array(cm_raw)
    if cm_array is not None:
        return json.dumps(cm_array)
    return None
```

```python
report_init["cm_json"] = report_init.confusion_matrix.apply(lambda x: convert_to_json(x))
report_init[["confusion_matrix", "cm_json"]]
```

```python
pd.DataFrame(json.loads(report_init.sample(1).cm_json.values[0]))
```
