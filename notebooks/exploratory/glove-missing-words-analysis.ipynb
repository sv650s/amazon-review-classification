# GloVe Missing Words Analysis

As part of model training, we place all words found in our corpus but did not find int he GloVe vocabulary list into files.

In this notebook we will look at missing vocab data

```python
# import sibling utilities
import sys
sys.path.append('../..')

import pandas as pd
from pandas import DataFrame
from pandas import Series
import numpy as np
import seaborn as sns
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(font_scale=2)
import logging
```

```python
missing_files = [
    'glove_embedding-missing_words-50k.csv',
    'glove_embedding-missing_words-100k.csv',
    'glove_embedding-missing_words-200k.csv',
    'glove_embedding-missing_words-500k.csv',
    'glove_embedding-missing_words-1m.csv',
    'glove_embedding-missing_words-2m.csv',
    'glove_embedding-missing_words-all.csv'
]

# key is number of samples, value is DF with missing words
missing_dict = {}
dict_keys = [50000, 100000, 200000, 500000, 1000000, 2000000, 9000000]
missing_wc = []
index = 0
for file in missing_files:
    df = pd.read_csv(f'../../reports/{file}')
    missing_dict[dict_keys[index]] = df
    missing_wc.append(len(df))
    index += 1
```

```python
f, a = plt.subplots(1, 1, figsize = (20, 5))
_ = sns.lineplot(x = dict_keys, y = missing_wc, marker = "o")
_ = plt.xlabel("Samples")
_ = plt.ylabel("Missing Words")
_ = plt.title("Missing Words by Sample Size")
```
