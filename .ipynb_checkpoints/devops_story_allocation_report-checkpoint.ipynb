# DevOps Story Allocation Report


Jira report requires the following fields:
* Issue key
* Issue type
* Epic Link
* Story Points
* Summary
* Resolved
* Status
* Resolution

Epic CVS file requires the following fields from Jira:
* Issue key
* Summary
* category1 - manually created for now
* category2 - manually created for now


DevOps Interruption file was manually manipulated to look like Jira report so that we can make merging easier

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime, timedelta, date

%matplotlib inline



# set global variables
story_date_columns = ["Created", "Resolved", "Updated"]
category_columns = ["Issue Type", "Status", "Resolution"]
drop_columns = ["Reporter", "Assignee", "Created", "Updated"]
c_issue_key_orig = "Issue key"
c_issue_key = "issue_key"
c_issue_type = "issue_type"
c_epic_id_orig = "Custom field (Epic Link)"
c_epic_id = "epic_id"
c_summary_orig = "Summary"
c_summary = "summary"
c_sp_orig = "Custom field (Story Points)"
c_sp = "story_points"
c_resolved = "resolved"
c_status = "status"
c_resolution = "resolution"
c_epic_name = "epic_name"
c_category1 = "category1"
c_category2 = "category2"
c_week = "week"

MINT = "MINT"
MINT_AVG = "MINT (Avg)"
DEVOPS = "DevOps"
DEVOPS_AVG = "DevOps (Avg)"
CAPPS = "CAPPS/DATA"
CAPPS_AVG = "CAPPS/DATA (Avg)"

# filters
COLUMN_FILTER = [c_issue_key, c_summary, c_issue_type, c_sp, c_epic_name, c_category1, c_category2, c_week]
ISSUE_TYPE_FILTER = ["Story", "Bug", "Task"]

# when calculating average - this is the window of weeks we are going to use
AVERAGE_WINDOW = 4
# weed out entries before this date for reporting purposes
START_DATE = pd.Timestamp(2019,1,31)

# output file names
OUT_MINT = f"{START_DATE.strftime('%Y-%m-%d')}-mint.csv"
OUT_DEVOPS = f"{START_DATE.strftime('%Y-%m-%d')}-devops.csv"
OUT_CAPPS = f"{START_DATE.strftime('%Y-%m-%d')}-capps_data.csv"
```

### Pre-processing

```python
story_csv = "data/201905-jira-report.csv"
epic_csv = "data/201905-epic_map.csv"
interruptions_csv = "data/201905-devops_interruptions.csv"

# load all DF's
story_df = pd.read_csv(story_csv, parse_dates=["Resolved"])
epic_df = pd.read_csv(epic_csv)
if interruptions_csv and len(interruptions_csv) > 0:
    interruptions_df = pd.read_csv(interruptions_csv, parse_dates=["Resolved"])
    # merge in interruptions
    merged_df = story_df.append(interruptions_df, ignore_index=True, sort=False)
# merged_df.tail(10)
# do epic lookups
merged_df = story_df.merge(epic_df[[c_issue_key_orig, c_summary_orig, c_category1, c_category2]], 
                           how="left", left_on=c_epic_id_orig, right_on=c_issue_key_orig)


# rename columns
column_mapper = {"Issue key_x": c_issue_key, 
                 "Summary_y": c_epic_name, 
                  "Summary_x": c_summary, 
                  "Custom field (Story Points)": c_sp,
                  "Custom field (Epic Link)": c_epic_id,
                 "Resolved": c_resolved,
                 "Status": c_status,
                 "Resolution": c_resolution,
                 "Issue Type": c_issue_type
                }
merged_df = merged_df.rename(column_mapper, axis=1)


# print(merged_df.columns)

# merged_df.head()

```

```python
# create data that we need
# add week column to group
df = merged_df

# create week so we can group by this
# TODO: when I convert it to start of the week - ie, date, the bar graphs do not plot this properly
df[c_week] = df[c_resolved].dt.week
# clean_df[c_week] = merged_df['Resolved'].dt.to_period('W').apply(lambda r: r.start_time)


# for Mint stories, put MINT in category1 and epic name in category2
# TODO: for CAPPS/DATA, this is set manually in the epic csv file manually - we should split out
# their own project to get more granular data
df.loc[df.issue_key.str.contains("MINTDEVO"), c_category1] = "MINT"
df.loc[df.category1 == "MINT", c_category2] = df[df.category1 == "MINT"].epic_name

# filtering

# filter out dates before our report
df = df[(df[c_resolved] >= START_DATE)]

# filter by issue type
df = df[df[c_issue_type].isin(ISSUE_TYPE_FILTER)]


# keep only if we have story points and we actually worked on it
df = df[(df.status == "Done") & (df.resolution == "Done") & (df.story_points.notnull()) & (df.story_points > 0)]

# filter column last so we can use all columns for filtering
df = df[COLUMN_FILTER]


# df.head(50)
```

# Overall Time Spent

* Not a significant difference when we look at story points vs number of stories ~ 10% difference
* Overall spending around 39 to 45% on DevOps
* Spending 30 to 40% on MINT team - target should be 17%

```python
# create df that has both story count and story points

pie_df1 = pd.DataFrame(df[[c_category1, c_sp]].groupby(c_category1).sum())
# print(pie_df1.head())
pie_df1 = pie_df1.merge(pd.DataFrame(df[[c_category1, c_sp]].groupby(c_category1).size()), left_index=True, right_index=True)
pie_df1 = pie_df1.rename({0: "count"}, axis=1)


plt.rcParams.update({'font.size': 22})
pie_df1.plot.pie(subplots=True, figsize=(30, 10), autopct='%1.1f%%')
```

```python
mint_results = df[df[c_category1] == MINT][[c_sp, c_category2]].groupby(c_category2)
mint_pie_df2 = mint_results.sum()
mint_pie_df2 = mint_pie_df2.merge(pd.DataFrame(mint_results.size()), left_index=True, right_index=True)
mint_pie_df2 = mint_pie_df2.rename({0: "count"}, axis=1)
# print(mint_pie_df2.head())

plt.rcParams.update({'font.size': 22})
mint_pie_df2.plot.pie(subplots=True, figsize=(30, 10), autopct='%1.1f%%')

df[df[c_category1] == MINT].to_csv(OUT_MINT)
```

### CAPPS/DATA Breakdown

```python
capps_results = df[df[c_category1] == CAPPS][[c_sp, c_category2]].groupby(c_category2)
capps_pie_df2 = capps_results.sum()
capps_pie_df2 = capps_pie_df2.merge(pd.DataFrame(capps_results.size()), left_index=True, right_index=True)
capps_pie_df2 = capps_pie_df2.rename({0: "count"}, axis=1)
# print(capps_pie_df2.head())

plt.rcParams.update({'font.size': 22})
capps_pie_df2.plot.pie(subplots=True, figsize=(30, 10), autopct='%1.1f%%')

# write to file
df[df[c_category1] == CAPPS].to_csv(OUT_CAPPS)
```

### DevOps Breakdown

```python
devops_results = df[df[c_category1] == DEVOPS][[c_sp, c_category2]].groupby(c_category2)
# print(devops_results)
devops_pie_df2 = devops_results.sum()
devops_pie_df2 = devops_pie_df2.merge(pd.DataFrame(devops_results.size()), left_index=True, right_index=True)
devops_pie_df2 = devops_pie_df2.rename({0: "count"}, axis=1)
# print(devops_pie_df2.head())

plt.rcParams.update({'font.size': 22})
devops_pie_df2.plot.pie(subplots=True, figsize=(30, 10), autopct='%1.1f%%')

# write to file
df[df[c_category1] == DEVOPS].to_csv(OUT_DEVOPS)
```

# Time Series Trending Data

```python
devops_df = df[df.category1 == DEVOPS][[c_week, c_sp]]
mint_df = df[df.category1 == MINT][[c_week, c_sp]]
team_df = df[df.category1 == CAPPS][[c_week, c_sp]]
```

```python
# Create table of story points
devops_sum = devops_df.groupby([c_week]).sum().fillna(0).reset_index()
mint_sum = mint_df.groupby([c_week]).sum().fillna(0).reset_index()
team_sum = team_df.groupby([c_week]).sum().fillna(0).reset_index()

# print(devops)
# print(mint)
# print(team)

table_sum = devops_sum.merge(mint_sum, how="left", left_on="week", right_on="week")
table_sum = table_sum.merge(team_sum, how="left", left_on="week", right_on="week")

# print(table)

table_sum = table_sum.rename({"story_points_x": DEVOPS, "story_points_y": MINT, "story_points": CAPPS}, axis=1)


# calculate averages
# add average columns
table_sum[DEVOPS_AVG] = round(table_sum[DEVOPS].fillna(0).rolling(window=AVERAGE_WINDOW).mean(),0)
table_sum[MINT_AVG] = round(table_sum[MINT].fillna(0).rolling(window=AVERAGE_WINDOW).mean(),0)
table_sum[CAPPS_AVG] = round(table_sum[CAPPS].fillna(0).rolling(window=AVERAGE_WINDOW).mean(),0)



table_sum = table_sum.fillna(0)

# print(table_sum)
# print(table_sum.shape)
```

```python
# Create table of story points
devops_count = devops_df.groupby([c_week]).size().fillna(0).reset_index()
mint_count = mint_df.groupby([c_week]).size().fillna(0).reset_index()
team_count = team_df.groupby([c_week]).size().fillna(0).reset_index()

# print(devops)
# print(mint)
# print(team)

table_count = devops_count.merge(mint_sum, how="left", left_on="week", right_on="week")
table_count = table_count.merge(team_sum, how="left", left_on="week", right_on="week")

# print(table)
table_count = table_count.rename({0: DEVOPS, "story_points_x": MINT, "story_points_y": CAPPS}, axis=1)


# calculate averages
# add average columns
table_count[DEVOPS_AVG] = round(table_count[DEVOPS].fillna(0).rolling(window=AVERAGE_WINDOW).mean(),0)
table_count[MINT_AVG] = round(table_count[MINT].fillna(0).rolling(window=AVERAGE_WINDOW).mean(),0)
table_count[CAPPS_AVG] = round(table_count[CAPPS].fillna(0).rolling(window=AVERAGE_WINDOW).mean(),0)



table_count = table_count.fillna(0)

# print(table_sum)
# print(table_sum.shape)
```

## Trending Data - Time spent on DevOps project decreasing over time - we are now at around 25% to 33% on DevOps projects and rest supporting other teams

```python
# Stacked Area Plot of Resolved Story Points

# if I want to plot both
f, ax = plt.subplots(2, 1, figsize=(20,20))

# plot rolling average
ax[0].set_title("Story Points By Week - Rolling Average")
ax[0].set_xlabel("Week")
ax[0].stackplot(table_sum[c_week], table_sum[DEVOPS_AVG].T, table_sum[MINT_AVG].T, table_sum[CAPPS_AVG].T,  
              labels=[DEVOPS, MINT, CAPPS], baseline="zero")
ax[0].legend(loc='upper right')
devops_rolling_avg = table_sum.iloc[table_sum.index.max()][DEVOPS_AVG]


total_rolling_avg = table_sum.iloc[table_sum.index.max()][DEVOPS_AVG] + \
                    table_sum.iloc[table_sum.index.max()][MINT_AVG] + \
                    table_sum.iloc[table_sum.index.max()][CAPPS_AVG]
ax[0].hlines(total_rolling_avg, table_sum.iloc[table_sum.index.min()][c_week] + 1, 
             table_sum.iloc[table_sum.index.max()][c_week], linestyles='dashed', label = f'{total_rolling_avg}')
ax[0].text(table_sum.iloc[table_sum.index.min()][c_week], total_rolling_avg, f'{total_rolling_avg}', ha="left", va="center")

ax[0].hlines(devops_rolling_avg, table_sum.iloc[table_sum.index.min()][c_week] + 1, 
             table_sum.iloc[table_sum.index.max()][c_week], linestyles='solid', label = f'{devops_rolling_avg}')
ax[0].text(table_sum.iloc[table_sum.index.min()][c_week], devops_rolling_avg, f'{devops_rolling_avg}', ha="left", va="center")


# plot rolling average
ax[1].set_title("Number of Stories By Week - Rolling Average")
ax[1].set_xlabel("Week")
ax[1].stackplot(table_count[c_week], table_count[DEVOPS_AVG].T, table_count[MINT_AVG].T, table_count[CAPPS_AVG].T,  
              labels=[DEVOPS, MINT, CAPPS], baseline="zero")
ax[1].legend(loc='upper right')

devops_rolling_avg = table_count.iloc[table_count.index.max()][DEVOPS_AVG]


total_rolling_avg = table_count.iloc[table_count.index.max()][DEVOPS_AVG] + \
                    table_count.iloc[table_count.index.max()][MINT_AVG] + \
                    table_count.iloc[table_count.index.max()][CAPPS_AVG]
ax[1].hlines(total_rolling_avg, table_count.iloc[table_count.index.min()][c_week] + 1, 
             table_count.iloc[table_count.index.max()][c_week], linestyles='dashed', label = f'{total_rolling_avg}')
ax[1].text(table_count.iloc[table_count.index.min()][c_week], total_rolling_avg, f'{total_rolling_avg}', ha="left", va="center")

ax[1].hlines(devops_rolling_avg, table_count.iloc[table_count.index.min()][c_week] + 1, 
             table_count.iloc[table_count.index.max()][c_week], linestyles='solid', label = f'{devops_rolling_avg}')
ax[1].text(table_count.iloc[table_count.index.min()][c_week], devops_rolling_avg, f'{devops_rolling_avg}', ha="left", va="center")



```
